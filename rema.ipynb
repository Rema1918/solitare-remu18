```python

!pip install gtts scikit-learn


from gtts import gTTS
from IPython.display import Audio, display
import random
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

intents = {
    "greet": {
        "patterns": ["hi", "hello", "hey", "good morning", "good evening"],
        "responses": ["Hello!", "Hi there!", "Hey! How can I help you?"]
    },
    "bye": {
        "patterns": ["bye", "goodbye", "see you later"],
        "responses": ["Goodbye!", "See you soon!", "Take care!"]
    },
    "thanks": {
        "patterns": ["thanks", "thank you", "thx"],
        "responses": ["You're welcome!", "No problem!", "Glad to help!"]
    },
    "about": {
        "patterns": ["who are you", "what are you", "introduce yourself"],
        "responses": ["I am a Colab chatbot.", "I'm your text-to-speech assistant!"]
    }
}

X_train, y_train = [], []
for intent, data in intents.items():
    for pattern in data["patterns"]:
        X_train.append(pattern)
        y_train.append(intent)

vectorizer = TfidfVectorizer()
X_vec = vectorizer.fit_transform(X_train)

model = MultinomialNB()
model.fit(X_vec, y_train)

def get_response(user_input):
    vec = vectorizer.transform([user_input.lower()])
    intent = model.predict(vec)[0]
    response = random.choice(intents[intent]["responses"])
    return response

def speak_response(response):
    """Convert text to speech using gTTS and play in Colab"""
    tts = gTTS(text=response, lang='en')
    tts.save("response.mp3")
    display(Audio("response.mp3", autoplay=True))

print("Colab Chatbot is running! Type 'quit' to exit.\n")

while True:
    user_input = input("You: ").strip()
    if not user_input:
        print("Bot: Please type something!")
        continue
    if user_input.lower() in ["quit", "exit"]:
        response = "Goodbye!"
        print("Bot:", response)
        speak_response(response)
        break
    response = get_response(user_input)
    print("Bot:", response)
    speak_response(response)
```
